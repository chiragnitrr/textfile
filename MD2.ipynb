- Code for Principal Stresses , Maximum Shear Stresses and the angles

```python
# Unit of stresses are in MPa
import math
sigma_xx = 50
sigma_yy = 30
tau_xy = 20

#Calculation of sigma1 , sigma2 and tau max
sigma_1 = (sigma_xx + sigma_yy)/2 + ( (sigma_xx - sigma_yy)**2/4 + tau_xy**2 )** 0.5
sigma_2 = (sigma_xx + sigma_yy)/2 - ( (sigma_xx - sigma_yy)**2/4 + tau_xy**2 )** 0.5

sigma_3 = 0

tau_max = max( abs((sigma_1 - sigma_2)/2), abs((sigma_2 - sigma_3)/2), abs((sigma_3 - sigma_1)/2) )

#Calculation for angles 
# tan(2*theta_1) = 2*tau_xy/(sigma_1 - sigma_2)
# theta_2 = theta_1 + 90 (degree)

theta_1 =  math.atan(2*tau_xy / (sigma_xx - sigma_yy)) /2
theta1 = math.degrees(theta_1)
theta2 = theta1 + 90

print("Printing the results")
print("Magnitude of the first principal stress : {:.3f} MPa".format(sigma_1))
print("Magnitude of the second principal stress : {:.3f} MPa".format(sigma_2))
print("Maximum shear stress : {:.3f} MPa".format(tau_max))
print("Angle of the first principal stress : {:.3f} degree".format(theta1))
print("Angle of the second principal stress : {:.3f} degree".format(theta2))
print("Maximum shear plane angle is : {:.3f} degrees, {:.3f} degree ".format(theta1 - 45, theta1 + 45))
```

- Code for Mohar's circle

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate centers of the circles
center_1 = sigma_1/ 2
center_2 = (sigma_2 + sigma_1)/2
center_3 = sigma_2/2

# Calculate radii of the circles
radius_1 = abs( (sigma_1 - sigma_3)/2 )
radius_2 = abs( (sigma_1 - sigma_2)/2 )
radius_3 = abs( (sigma_2 - sigma_3)/2 )

# Create an array of angles
theta = np.linspace(0, 2 * np.pi, 100)

# Calculate coordinates for the circles
x_1 = center_1 + radius_1 * np.cos(theta)
y_1 = radius_1 * np.sin(theta)

x_2 = center_2 + radius_2 * np.cos(theta)
y_2 = radius_2 * np.sin(theta)

x_3 = center_3 + radius_3 * np.cos(theta)
y_3 = radius_3 * np.sin(theta)

# Plotting the circles
plt.figure(figsize=(8, 6))
plt.plot(x_1, y_1, label='Mohr\'s Circle 1')
plt.plot(center_1, 0, 'ro', label='Center 1 of Circle')

plt.plot(x_2, y_2, label='Mohr\'s Circle 2')
plt.plot(center_2, 0, 'go', label='Center 2 of Circle')

plt.plot(x_3, y_3, label='Mohr\'s Circle 3')
plt.plot(center_3, 0, 'bo', label='Center 3 of Circle')

plt.axhline(y=0, color='k', linestyle='--', label='Axis of Maximum Shear Stress')

# Adding labels and legend
plt.xlabel('Normal Stress')
plt.ylabel('Shear Stress')
plt.title('Mohr\'s Circles')
plt.grid()
plt.legend()
plt.axis('equal')

# Show the plot
plt.show()
```
